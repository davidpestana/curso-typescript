## üß© Ejercicio 3 ‚Äî Narrowing con `in`

### üéØ Reto
Define dos tipos:

```ts
type Admin = { rol: "admin", permisos: string[] }
type Invitado = { rol: "invitado", visitas: number }
```

Crea una funci√≥n `infoUsuario` que reciba `Admin | Invitado` y determine qu√© tipo es usando `in`.

````typescript
// ‚úèÔ∏è Tu soluci√≥n
// function infoUsuario(u: Admin | Invitado) { ... }
````

### ‚úÖ Soluci√≥n Ejercicio 3

````typescript
type Admin = { rol: "admin", permisos: string[] };
type Invitado = { rol: "invitado", visitas: number };

function infoUsuario(u: Admin | Invitado) {
  if ("permisos" in u) {
    console.log("Es admin ‚Üí", u.permisos);
  } else {
    console.log("Es invitado ‚Üí visitas:", u.visitas);
  }
}

infoUsuario({ rol: "admin", permisos: ["crear", "borrar"] });
infoUsuario({ rol: "invitado", visitas: 4 });
````

## üß© Ejercicio 5 ‚Äî Discriminated Union

### üéØ Reto
Modela veh√≠culos:

```ts
type Coche = { tipo: "coche", puertas: number }
type Moto = { tipo: "moto", cilindrada: number }
```

Crea una funci√≥n `infoVehiculo(v)` que use `switch` sobre `v.tipo`.

````typescript
// ‚úèÔ∏è Tu soluci√≥n
// function infoVehiculo(v: Coche | Moto) { ... }
````

### ‚úÖ Soluci√≥n Ejercicio 5

````typescript
type Coche = { tipo: "coche", puertas: number };
type Moto = { tipo: "moto", cilindrada: number };

function infoVehiculo(v: Coche | Moto) {
  switch (v.tipo) {
    case "coche":
      console.log("Coche con puertas ‚Üí", v.puertas);
      break;
    case "moto":
      console.log("Moto con cilindrada ‚Üí", v.cilindrada);
      break;
  }
}

infoVehiculo({ tipo: "coche", puertas: 5 });
infoVehiculo({ tipo: "moto", cilindrada: 600 });
````

## üß© Ejercicio 6 ‚Äî `never` y exhaustividad

### üéØ Reto
Implementa una funci√≥n `area(fig)` usando un `switch` exhaustivo.

Si aparece un tipo no manejado, debe llamar a una funci√≥n que reciba `never`.

````typescript
// ‚úèÔ∏è Tu soluci√≥n
// function assertNever(x: never): never { ... }
// function area(fig: Figura) { ... }
````

### ‚úÖ Soluci√≥n Ejercicio 6

````typescript
type Cuadrado = { tipo: "cuadrado", lado: number };
type Circulo = { tipo: "circulo", radio: number };
type Figura = Cuadrado | Circulo;

function assertNever(x: never): never {
  throw new Error("Caso no manejado: " + x);
}

function area(fig: Figura) {
  switch (fig.tipo) {
    case "cuadrado":
      return fig.lado ** 2;
    case "circulo":
      return Math.PI * fig.radio ** 2;
    default:
      assertNever(fig);
  }
}

console.log(area({ tipo: "cuadrado", lado: 5 }));
console.log(area({ tipo: "circulo", radio: 3 }));
````
