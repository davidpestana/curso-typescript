## üß© Ejercicio 2 ‚Äî Propiedades privadas
Crea una clase `Cuenta` que tenga:
- `numero: string` (read-only)
- `saldo: number` (privado)

Debe tener los m√©todos:
- `depositar(cantidad)`
- `retirar(cantidad)` (solo si hay saldo)
- `verSaldo()`

```typescript
// Tu soluci√≥n aqu√≠
```

```typescript
// Soluci√≥n
class Cuenta {
  private saldo: number = 0;
  constructor(public readonly numero: string) {}

  depositar(m: number) {
    this.saldo += m;
  }
  retirar(m: number) {
    if (m <= this.saldo) this.saldo -= m;
  }
  verSaldo() {
    return this.saldo;
  }
}

const c = new Cuenta("ES001");
c.depositar(200);
c.retirar(50);
console.log(c.verSaldo());
```

## üß© Ejercicio 4 ‚Äî M√©todos est√°ticos
Crea una clase `MathUtil` con:
- m√©todo est√°tico `aleatorio()` que devuelva un n√∫mero entre 1 y 100
- m√©todo est√°tico `doble(n)` que devuelva `n * 2`

```typescript
// Tu soluci√≥n aqu√≠
```

```typescript
// Soluci√≥n
class MathUtil {
  static aleatorio() {
    return Math.floor(Math.random() * 100) + 1;
  }
  static doble(n: number) { return n * 2; }
}

console.log(MathUtil.aleatorio());
console.log(MathUtil.doble(4));
```
