### üß™ Ejercicio

### ‚úÖ Soluci√≥n Paso 6

````typescript
let entrada: unknown = { nombre: "Batidora", precio: 49.99, categoria: "HOGAR" };

function esProducto(x: unknown): x is Producto {
  if (typeof x === "object" && x !== null) {
    const asObj = x as any;
    return (
      typeof asObj.nombre === "string" &&
      typeof asObj.precio === "number" &&
      typeof asObj.categoria === "string"
    );
  }
  return false;
}

if (esProducto(entrada)) {
  console.log("Entrada v√°lida como Producto ‚Üí", entrada);
} else {
  console.log("‚ùå Entrada inv√°lida");
}
````
