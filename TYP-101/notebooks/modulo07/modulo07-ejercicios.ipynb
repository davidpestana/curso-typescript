## üß© Ejercicio 1 ‚Äî Herencia b√°sica

Crea una clase base `Persona` con:
- propiedad `nombre`
- m√©todo `saludar()` que muestre: "Hola, soy <nombre>"

Crea una subclase `Estudiante` que extienda `Persona` y a√±ada:
- propiedad `curso`
- m√©todo `info()` que muestre: "Estudio <curso>"

````typescript
// ‚úèÔ∏è Tu soluci√≥n al Ejercicio 1
````

````typescript
// ‚úÖ Soluci√≥n
class Persona {
  constructor(public nombre: string) {}
  saludar() {
    console.log(`Hola, soy ${this.nombre}`);
  }
}

class Estudiante extends Persona {
  constructor(nombre: string, public curso: string) {
    super(nombre);
  }
  info() {
    console.log(`Estudio ${this.curso}`);
  }
}

const e1 = new Estudiante("Laura", "Matem√°ticas");
e1.saludar();
e1.info();
````

- Clase `Moto`, que extienda `Vehiculo`, con propiedad `cilindrada`

````typescript
// ‚úÖ Soluci√≥n
class Vehiculo {
  constructor(public marca: string) {}
}

class Moto extends Vehiculo {
  constructor(marca: string, public cilindrada: number) {
    super(marca);
  }
}

const m = new Moto("Honda", 125);
console.log(m.marca, m.cilindrada);
````

## üß© Ejercicio 3 ‚Äî Sobrescritura (`override`)

Crea:
- Clase base `Trabajador` con m√©todo `trabajar()` que imprima "Estoy trabajando"
- Clase `Programador` que sobrescriba `trabajar()` mostrando "Estoy programando"


````typescript
// ‚úèÔ∏è Tu soluci√≥n al Ejercicio 3
````

````typescript
// ‚úÖ Soluci√≥n
class Trabajador {
  trabajar() {
    console.log("Estoy trabajando");
  }
}

class Programador extends Trabajador {
  override trabajar() {
    console.log("Estoy programando");
  }
}

new Programador().trabajar();
````

## üß© Ejercicio 4 ‚Äî Clase abstracta

Crea una clase abstracta `Figura` con:
- m√©todo abstracto `area(): number`
- m√©todo normal `descripcion()` que diga "Soy una figura"

Crea dos clases hijas:
- `Cuadrado` con propiedad `lado`
- `Circulo` con propiedad `radio`

Ambas deben implementar `area()`.

````typescript
// ‚úèÔ∏è Tu soluci√≥n al Ejercicio 4
````

````typescript
// ‚úÖ Soluci√≥n
abstract class Figura {
  abstract area(): number;
  descripcion() { console.log("Soy una figura"); }
}

class Cuadrado extends Figura {
  constructor(public lado: number) { super(); }
  area() { return this.lado * this.lado; }
}

class Circulo extends Figura {
  constructor(public radio: number) { super(); }
  area() { return Math.PI * this.radio ** 2; }
}

console.log(new Cuadrado(3).area());
console.log(new Circulo(2).area());
````
