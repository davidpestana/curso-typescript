# üß© Reto 2 ‚Äî Clase `Jefe`

Crea la clase `Jefe` que extienda `Empleado`, a√±adiendo:
- propiedad `equipo: number` (n¬∫ de personas a su cargo)

**Regla salarial:**

```ts
salario = salarioBase + (equipo * 100)
```

````typescript
// ‚úèÔ∏è Tu soluci√≥n ‚Äî Reto 2
````

````typescript
// ‚úÖ Soluci√≥n ‚Äî Reto 2
class Jefe extends Empleado {
  constructor(nombre: string, salarioBase: number, public equipo: number) {
    super(nombre, salarioBase);
  }

  override calcularSalario() {
    return this.salarioBase + this.equipo * 100;
  }
}

const j = new Jefe("Carlos", 2000, 5);
j.info();
````

# üß© Reto 3 ‚Äî Clase `Director`

Crea la clase `Director`, que hereda de `Jefe`, a√±adiendo:

- `departamentos: number` (cu√°ntos departamentos gestiona)

**Regla salarial:**

```ts
salario = salarioBase
         + (equipo * 100)
         + (departamentos * 500)
```

````typescript
// ‚úèÔ∏è Tu soluci√≥n ‚Äî Reto 3
````

````typescript
// ‚úÖ Soluci√≥n ‚Äî Reto 3
class Director extends Jefe {
  constructor(nombre: string, salarioBase: number, equipo: number, public departamentos: number) {
    super(nombre, salarioBase, equipo);
  }

  override calcularSalario() {
    return super.calcularSalario() + this.departamentos * 500;
  }
}

const d = new Director("Mar√≠a", 3000, 10, 3);
d.info();
````

# üß© Reto 4 ‚Äî Polimorfismo


````typescript
// ‚úèÔ∏è Tu soluci√≥n ‚Äî Reto 4
````

````typescript
// ‚úÖ Soluci√≥n ‚Äî Reto 4
function imprimirNominas(empleados: Empleado[]) {
  console.log("--- N√≥minas ---");
  for (const e of empleados) {
    e.info();
  }
}

imprimirNominas([
  new Jefe("Carlos", 2000, 5),
  new Director("Mar√≠a", 3000, 10, 3)
]);
````
