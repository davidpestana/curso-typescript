# ðŸ§© Ejercicio 3 â€” ParÃ¡metros rest
Crea una funciÃ³n `media(...nums: number[]): number` que calcule la media.

```typescript
// Tu soluciÃ³n
```

```typescript
// SoluciÃ³n
function media(...nums: number[]): number {
  return nums.reduce((acc, n) => acc + n, 0) / nums.length;
}
console.log(media(5, 10, 15));
```

# ðŸ§© Ejercicio 4 â€” Call signature en interfaz

```typescript
// Tu soluciÃ³n
```

```typescript
// SoluciÃ³n
interface Comparador {
  (a: number, b: number): boolean;
}
const mayorQue: Comparador = (a, b) => a > b;
console.log(mayorQue(10, 5));
```
