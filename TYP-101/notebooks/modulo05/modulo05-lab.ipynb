# üß© Reto 1 ‚Äî Implementar `map` tipado

Crea una funci√≥n:

`map<T, U>(arr: T[], fn: (item: T) => U): U[]`

Debe aplicar la funci√≥n a cada elemento y devolver un nuevo array.

````typescript
// Tu soluci√≥n aqu√≠
````

````typescript
// Soluci√≥n
function map<T, U>(arr: T[], fn: (item: T) => U): U[] {
  const resultado: U[] = [];
  for (const item of arr) {
    resultado.push(fn(item));
  }
  return resultado;
}

console.log(map([1, 2, 3], x => x * 10));
````

# üß© Reto 2 ‚Äî Implementar `filter` tipado

Crea una funci√≥n:

`filter<T>(arr: T[], fn: (item: T) => boolean): T[]`

Debe devolver un array con los elementos que cumplan la condici√≥n.

````typescript
// Tu soluci√≥n aqu√≠
````

````typescript
// Soluci√≥n
function filter<T>(arr: T[], fn: (item: T) => boolean): T[] {
  const resultado: T[] = [];
  for (const item of arr) {
    if (fn(item)) resultado.push(item);
  }
  return resultado;
}

console.log(filter([1, 2, 3, 4, 5], n => n % 2 === 0));
````

# üß© Reto 3 ‚Äî Implementar `reduce` tipado

Crea una funci√≥n:

`reduce<T, U>(arr: T[], fn: (acc: U, item: T) => U, inicial: U): U`

Debe acumular los valores seg√∫n la funci√≥n recibida.

````typescript
// Tu soluci√≥n aqu√≠
````

````typescript
// Soluci√≥n
function reduce<T, U>(arr: T[], fn: (acc: U, item: T) => U, inicial: U): U {
  let acc = inicial;
  for (const item of arr) {
    acc = fn(acc, item);
  }
  return acc;
}

console.log(reduce([1, 2, 3, 4], (acc, x) => acc + x, 0));
````

# üß© Reto 4 ‚Äî Reutilizar las funciones

Usa tus funciones `map`, `filter` y `reduce` para:

1. Filtrar los n√∫meros pares de `[1, 2, 3, 4, 5, 6]`
2. Multiplicarlos √ó10
3. Sumar el resultado final

Todo encadenando funciones.

````typescript
// Tu soluci√≥n aqu√≠
````

````typescript
// Soluci√≥n
const datos = [1, 2, 3, 4, 5, 6];

const pares = filter(datos, n => n % 2 === 0);
const multiplicados = map(pares, n => n * 10);
const suma = reduce(multiplicados, (acc, x) => acc + x, 0);

console.log({ pares, multiplicados, suma });
````

# üß© Reto 5 ‚Äî Mini Result<T>

Crea un tipo simple `Result` basado en funciones avanzadas:

```ts
type Ok<T> = { ok: true; value: T }
type Err = { ok: false; error: string }

type Result<T> = Ok<T> | Err
```

Y una funci√≥n `safeDivide(a: number, b: number): Result<number>` que:

- Si `b = 0`, devuelva `{ ok: false, error: "No se puede dividir" }`
- Si no, devuelva `{ ok: true, value: a / b }`

````typescript
// Tu soluci√≥n aqu√≠
````

````typescript
// Soluci√≥n
type Ok<T> = { ok: true; value: T };
type Err = { ok: false; error: string };
type Result<T> = Ok<T> | Err;

function safeDivide(a: number, b: number): Result<number> {
  if (b === 0) return { ok: false, error: "No se puede dividir" };
  return { ok: true, value: a / b };
}

console.log(safeDivide(10, 2));
console.log(safeDivide(10, 0));
````
